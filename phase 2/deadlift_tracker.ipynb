```python
import pandas as pd
import pickle

cols = []
for i in range(0, 33):
    cols += [f'x{i}', f'y{i}', f'z{i}', f'v{i}']
temp = pd.DataFrame(columns=cols)
```

```python
loaded_model = pickle.load(open('deadlift_model.sav', 'rb'))
```

```python
import cv2
import mediapipe as mp
from google.protobuf.json_format import MessageToDict
import numpy as np
import xgboost


mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

```python
def predictor(landmarks):
    data_list = []
    for item in landmarks:
        dict_item = MessageToDict(item)
        data_list.append(dict_item)

    # Create a Pandas dataframe from the list of dictionaries
    df = pd.DataFrame(data_list)
    data = df.values.flatten()
    temp.loc[len(df)] = data
    y_pred = loaded_model.predict(temp)
    
    if y_pred == 1:
        Class = 'up'
    elif y_pred == 0:
        Class = 'mid'
    elif y_pred == -1:
        Class ='down'
    return(Class)
    
```

```python
import cv2
import mediapipe as mp
import pandas as pd
from google.protobuf.json_format import MessageToDict


mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_pose = mp.solutions.pose

cap = cv2.VideoCapture('deadlift_test.mov')
with mp_pose.Pose(min_detection_confidence=0.5,min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        success, image = cap.read()
        if not success:
              print("Ignoring empty camera frame.")
          # If loading a video, use 'break' instead of 'continue'.
              break

        # To improve performance, optionally mark the image as not writeable to
        # pass by reference.
        image.flags.writeable = False
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        results = pose.process(image)

        # Draw the pose annotation on the image.
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        mp_drawing.draw_landmarks(
            image,
            results.pose_landmarks,
            mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing_styles.get_default_pose_landmarks_style())
        # Flip the image horizontally for a selfie-view display.
        cv2.imshow('MediaPipe Pose', cv2.flip(image, 0))
        
        
        
        
        landmarks = results.pose_landmarks.landmark
        pos = predictor(landmarks)
        print(pos)
        
        
        
        
        if cv2.waitKey(5) & 0xFF == 27:
              break
cap.release()
cv2.destroyAllWindows()
```
